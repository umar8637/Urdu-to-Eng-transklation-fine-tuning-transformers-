```python
dataset= "/kaggle/input/urdu-to-english-dataset/Dataset.txt"
```

```python
file= open(dataset, 'r', encoding='utf-8')
raw_data= file.read()

file.close()
```

```python
raw_data= raw_data.split('\n')
urdu_english_pairs= [sent.split('\t') for sent in raw_data]
```

```python
# Define the datasets without overlap
training_dataset = urdu_english_pairs[:1000]
validation_dataset = urdu_english_pairs[1000:1100]
testing_dataset = urdu_english_pairs[1100:]
```

```python

# Convert the datasets to dictionaries
training_dict = [{'ur': i[1], 'en': i[0]} for i in training_dataset]
validation_dict = [{'ur': i[1], 'en': i[0]} for i in validation_dataset]
testing_dict = [{'ur': i[1], 'en': i[0]} for i in testing_dataset]
```

```python
train_dict= {'translation': training_dict}
valid_dict= {'translation': validation_dict}
test_dict= {'translation': testing_dict}
```

```python
urdu_english_pairs
```

```python
from datasets import Dataset

train_dataset= Dataset.from_dict(train_dict)
valid_dataset= Dataset.from_dict(valid_dict)
test_dataset= Dataset.from_dict(test_dict)
```

```python
import datasets

final_dataset=datasets.DatasetDict(
    {
        'train': train_dataset,
        'validation': valid_dataset,
        'test': test_dataset
    }
)
```

```python
final_dataset
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
tokenizer= AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-ur-en")
```

```python
model= AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-ur-en")
```

```python
max_length= 256

def preprocess_function(examples):
    inputs= [ex['ur']for ex in examples['translation']]
    targets= [ex['en']for ex in examples['translation']]

    model_inputs= tokenizer(inputs, max_length=max_length, truncation=True)
    labels= tokenizer(targets, max_length= max_length, truncation=True)

    model_inputs['labels']= labels['input_ids']

    return model_inputs
```

```python
tokenized_dataset_train= final_dataset['train'].map(
    preprocess_function,
    batched= True,
    remove_columns= final_dataset['train'].column_names,
    batch_size=2
)
```

```python
tokenized_dataset_validation= final_dataset['validation'].map(
    preprocess_function,
    batched= True,
    remove_columns= final_dataset['validation'].column_names,
    batch_size=2
)
```

```python
tokenized_dataset_test= final_dataset['test'].map(
    preprocess_function,
    batched= True,
    remove_columns= final_dataset['test'].column_names,
    batch_size=2
)
```

```python
from transformers import DataCollatorForSeq2Seq

data_collator= DataCollatorForSeq2Seq(tokenizer, model= model)
```

```python
model_config= model.config
print(model_config)
```

```python
print(f"NUmber of encoder Layers are: {model_config.encoder_layers}")
print(f"Number of decoder Layers are: {model_config.decoder_layers}")
```

```python
for parameter in model.parameters():
    parameter.requires_grad= True

num_layers_to_freeze= 3

encoder= model.get_encoder()
for layer_index, layer in enumerate(encoder.layers):
    if layer_index < num_layers_to_freeze:
        for parameter in layer.parameters():
            parameter.requires_grad= False


decoder= model.get_decoder()
for layer_index, layer in enumerate(decoder.layers):
    if layer_index < num_layers_to_freeze:
        for parameter in layer.parameters():
            parameter.requires_grad= False
```

```python
print("Encoder layers status:")
for i, layer in enumerate(encoder.layers):
    frozen = all(not param.requires_grad for param in layer.parameters())
    print(f"Encoder Layer {i} frozen: {frozen}")

print("Decoder layers status:")
for i, layer in enumerate(decoder.layers):
    frozen = all(not param.requires_grad for param in layer.parameters())
    print(f"Decoder Layer {i} frozen: {frozen}")
```

```python
import evaluate

metric= evaluate.load('sacrebleu')
```

```python
metric
```

```python
import numpy as np

def compute_metrics(eval_pred):
    preds, labels= eval_pred

    if isinstance(preds, tuple):
        preds= preds[0]

    decoded_preds= tokenizer.batch_decode(preds, skip_special_tokens=True)

    labels= np.where(labels !=-100, labels, tokenizer.pad_token_id)

    decoded_labels= tokenizer.batch_decode(labels, skip_special_tokens=True)

    decoded_preds= [pred.strip() for pred in decoded_preds]
    decoded_labels= [[label.strip()] for label in decoded_labels]

    result= metric.compute(predictions= decoded_preds, references=decoded_labels)

    return {'bleu: ': result['score']}
```

```python
import torch

device= torch.device('cuda' if torch.cuda.is_available() else 'cpu')

if torch.cuda.is_available():
    print('CUDA available, using GPU')
else:
    print('CUDA is not available, using CPU')
```

```python
model.to(device)
```

```python
from transformers import Seq2SeqTrainingArguments

training_args = Seq2SeqTrainingArguments(
    output_dir='./results',
    gradient_checkpointing=True,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    learning_rate=1e-5,
    warmup_steps=100,
    max_steps=3000,
    fp16=True,
    optim='adafactor',
    metric_for_best_model='eval_bleu: ',
    predict_with_generate=True,
    push_to_hub=False,
    num_train_epochs=10,
    logging_dir='./logs',
    save_total_limit=3,
    evaluation_strategy='steps',
    eval_steps=500,
    save_steps=500
)




```

```python
from transformers import Seq2SeqTrainer

trainer= Seq2SeqTrainer(
    model,
    training_args,
    train_dataset= tokenized_dataset_train,
    eval_dataset= tokenized_dataset_validation,
    data_collator= data_collator,
    tokenizer= tokenizer,
    compute_metrics= compute_metrics
)
```

```python
trainer.train()
```

```python
trainer.save_model('./my_model')  # Saves model weights and configuration
tokenizer.save_pretrained('./my_model')  # Saves tokenizer configuration
```
